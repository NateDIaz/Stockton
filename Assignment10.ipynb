````r
---
title: "Assignment 10"
author: "Rodrigue Beleho"
date: "October 15, 2018"
output: word_document
---

#Question 2
```{r setup, include=FALSE}
df = read.csv("adult.csv", header = FALSE)
```


#Question 3
```{r}
colnames(df)[1] <- "age"
colnames(df)[2] <- "sector"
colnames(df)[3] <- "IDnumber"
colnames(df)[4] <- "education_level"
colnames(df)[5] <- "education_number"
colnames(df)[6] <- "marital_status"
colnames(df)[7] <- "profession"
colnames(df)[8] <- "family_background"
colnames(df)[9] <- "race"
colnames(df)[10] <- "gender"
colnames(df)[11] <- "capital_gain"
colnames(df)[12] <- "capital_loss"
colnames(df)[13] <- "weekly_hours"
colnames(df)[14] <- "country"
colnames(df)[15] <- "salary"
```


#Question 4
```{r}
summary(df)
sum(is.na(df))

```


#Question 5
```{r}

for (i in 1:ncol(df)){
    df[df == "?"] = NA
    blanks <- sum(is.na(df[i])) 
    if (blanks >0){
      if(is.numeric(df[[i]]) == TRUE){
        df[[i]][is.na(df[[i]])] <- mean(df[[i]], na.rm = TRUE) } 
  
  else {
      levels=unique(df[,i])
      df[,i][is.na(df[,i])]=levels[which.max(tabulate(match(df[,i], x=  levels)))]
    }
    }
}

```


#Question 6
```{r}
library(ggplot2)
ggplot(data=df) + geom_bar(mapping = aes(x = salary, fill = education_level)) +facet_wrap(~education_level)

ggplot(data = df) + geom_density(mapping=aes(x = capital_loss, fill = gender)) + facet_wrap(~gender)

ggplot(data = df) + geom_density(mapping = aes(x = age, fill = race)) + facet_wrap(~race)

ggplot(data = df) + geom_bar(mapping = aes(x = race , fill = gender), position = "dodge")

```


#Question 7
```{r} 
library(caret)
names(df)[15] = "objective"
woodley <- createDataPartition(df$objective, p = 0.6, list = FALSE, times = 1)
#splitIndex = woodley

training <- df[woodley,]
testing <- df[-woodley,]


```


#Question 8
```{r}
library(rpart)
tree <- rpart(objective ~., data = training, method = "class")
prediction <- predict(tree,testing, type = "class")
```

#Question 9
```{r}
cm = confusionMatrix( data = prediction, reference= testing$objective, positive= " >50K")

print(cm)
```


#Question 10
```{r}
library(rattle)
fancyRpartPlot(tree)
```


#Question 11
```{r}

tree_model <- train(objective~.,data = training, method = "rpart")
prediction2 = predict(tree_model,testing)

cm2 = confusionMatrix(prediction2, testing$objective, positive= " >50K")
print(cm2)
```


#Question 12
```{r}
finalfunction <- function(df,trainpct){
  library(caret)
  library(rpart)
  library(rattle)
  
  woodley <- createDataPartition(df$objective, p = trainpct, list = FALSE, times = 1)
  training <- df[woodley,]
  testing <- df[-woodley,]
  tree <- rpart(objective ~ ., data = training, method = "class")
  prediction <- predict(tree,testing, type = "class")
  cm = confusionMatrix(data = prediction, reference = testing$objective, positive= " >50K")
  print(cm)
  fancyRpartPlot(tree)
}

finalfunction(df,0.7)

```
````
