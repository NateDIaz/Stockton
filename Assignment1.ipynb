````r


#Graph1

```{r}
titanic = read.csv("titanic.csv")
library(ggplot2)
ggplot(titanic) + geom_bar(mapping = aes(x = Pclass, fill = Sex), position = "dodge")

```

This graph shows that there are more males than females on average. Especially in class 3 where the number is almost double.

#Graph2

```{r}
ggplot(titanic) + geom_bar(mapping = aes(x = factor(Survived), fill = Sex), position = "dodge")
```

This graph shows that a lot of male died (over 5 times the amount of female) and more female than male survived.


#Graph3

```{r, warning=FALSE}
ggplot(titanic) + geom_boxplot(mapping = aes(y = Age))
```

This graph shows that 50% of the people on the Titanic were aged between 20 and 40.

#Graph4

```{r, warning=FALSE}
ggplot(titanic) + geom_density(mapping =  aes(x = Age, color = factor(Survived)))+facet_wrap(~ Pclass, nrow = 2)
```

This graph shows that on average younger people composed the group of people with the most survivor. It also tells us that younger people composed the group of people with the most deaths.


#Graph5

```{r}
ggplot(titanic) + geom_bar(mapping = aes(x = Sex, fill = Age), position = "dodge")
```
This graph shows that on the titanic, there are more males than females.


#Graph6

```{r, warning= FALSE}
ggplot(titanic) + geom_histogram(mapping = aes(x = Age, fill = Sex), position = "dodge")
```

This graph is a very precise one; It shows the number of people on the  Titanic by every single group of age (20,21,22, and so on) and by gender. For most cases, there are significantly more males than females.


#Graph7

```{r, warning= FALSE}
ggplot(titanic) + geom_histogram(mapping = aes(x = Age, fill = factor(Survived)), position = "dodge")+facet_wrap(~ Pclass, nrow = 2)
```

This graph tells us that on average people that were in class 1 have a higher chance of survival than people in other classes.People in class 3 have a lower chance of survival than any other group.



#Graph8
```{r, warning=FALSE}
ggplot(data = titanic) + 
  geom_smooth(mapping = aes(x = Fare, y = Age, linetype = factor(Survived)))
```
This graph shows that the more people pay for their tickets, the higher they have chances to survive.


#Graph9 
```{r}
ggplot(titanic) + geom_density(mapping =  aes(x = Age, color = factor(Sex)))+facet_wrap(~ Pclass, nrow = 2)
```

This graph shows that for class 1 and 2, there are approcimately equally the amount of make and female. In class 3 however, there are more males than females.
````
