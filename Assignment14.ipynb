```r
#' Question 1 - Quickclean function

#' The goal of this function is to handle missing data by
#' removing observations with missing values or
#' imputing the missing values by the mean/mode of the corresponding columns.


Quickclean = function(x,i)
{
  if(i==0)
  {
    x = x[complete.cases(x), ]
    return(x)
  }
  if(i==1)
  {
    checkcol = function(y)
    {
      if(is.numeric(y) == TRUE)
      {
        y[is.na(y)] = mean(y, na.rm = TRUE)
      }
      else
      {
        levels=unique(y)
        y[is.na(y)]=levels[which.max(tabulate(match(y, x = levels)))]
      }
      x = data.frame(lapply(x,checkcol))
      return(x)
    }
  }
}


# Question 2 - quickvisual
# The goal of the function is to create some graphs: bar charts density plots,
# scatter plots with all possible combinations of variables.

#' @param df is a dataframe
#' @param task specifies the visualization to produce.
#' 1: plot the possible barcharts of 2 small categorical variables
#' 2: plot all density curves of numeric variables partioning on each small categorical variables
#' 3: plot scatter plot of all possible of all possible plair of numeric variable with less than 5

quickvisual <- function(df, task)
{
  library(ggplot2)

  if(is.data.frame(df) & task == 1){

  df_factors <- df[,sapply(df, is.factor)]

  cat.5 <- function(x)
  {
    l <- (length(unique(x)) < 5)
    return(l)
  }

  df_factors5 <- df_factors[,sapply(df_factors, cat.5)]
  combinations <- combn(names(df_factors5), 2, simplify = TRUE)

  bar.charts <- function(d)
  {
    chart <- ggplot(d) + geom_bar(mapping = aes(x=d[,1], fill=d[,2]), position = "dodge")+
    xlab(names(d[1])) + labs(fill = names(d[2]))
    return(chart)
  }

  c <- function(x)
  {
    bar.charts(df_factors5[,x])
  }

  charts <- apply(combinations, 2,  c)

  return(charts)
}


  if(is.data.frame(df) & task == 2)
    {


  x <- df

  for(i in 1:(ncol(x)-1)) {

    for(j in (i+1):ncol(x)) {

    print(ggplot(data=x)+geom_density(mapping=aes(x=x[,i]),fill = x[,j]))+xlab(names(x)[i])
                            }
                          }
    }


  if(is.data.frame(df) & task == 3)
  {

      df_numeric <- df[,sapply(df, is.numeric)]

    combinations <- combn(names(df_numeric), 2, simplify = TRUE)

    bar.charts <- function(d)
    {
      chart <- ggplot(d) + geom_point(mapping = aes(x=d[,1], y=d[,2])) +
      xlab(names(d[[1]])) + labs(fill = names(d[[2]]))
      return(chart)
    }

    c <- function(x)
    {
      bar.charts(df_numeric[,x])
    }

    charts <- apply(combinations, 2,  c)

    return(charts)
  }
}

#' Question 3 - Quickmodel with random forest
#'
#' The goal of this function is to
#' build a random forest using ranger package and output the accuracy


quickmodel = function(x,i)

{

  library(ranger)

  library(caret)

  colnames(x)[i] = "target"

  Grid = expand.grid(mtry = 2, splitrule = c("gini"), min.node.size = c(1:3))

  model = train(target~., x, method = "ranger", trControl = trainControl(method = "cv", number = 7), tuneGrid = Grid)

  return(max(model$results$Accuracy))

}
```
